```python
x=Hd.drop('target',axis=1)
y=Hd['target']
```

```python
x.shape
```

```python
y.shape
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier()
```

```python
DT=DT.fit(x_train,y_train)
```

```python
pred=DT.predict(x_test)
```

```python
print(pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('accuracy '),accuracy_score(pred,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
score=cross_val_score(dtc,x,y,cv=5,scoring='f1_macro')
print('mean score',score.mean())
print('standard deviation ',score.std() )
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print('confusion matrix ',confusion_matrix(pred,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
gnb.fit(x_train, y_train) 
```

```python
pred=gnb.predict(x_test)
print(' prediction \n',pred)
print('accuracy score\n',accuracy_score(pred,y_test))
print('confusion metrix\n',confusion_matrix(pred,y_test))
#print('classification Report \n',classification_report(pred,y_test))
```
